```python
import os
def get_dir(dir):
    p = []
    for root, dirs, files in os.walk(dir, topdown=False):
        for name in files:
                if name != '.DS_Store':
                    if not ('ipynb' in name):
                        p.append(os.path.join(root, name))
    return p
```

```python
##Outliers on age
```

```python
def outlier_threshold(df,x):
    q1 = df[x].quantile(0.3)
    q3 = df[x].quantile(0.75)
    IQR = q3-q1
    low  = q1 - 1.5*IQR
    high = q3 + 1.5*IQR
    return low, high
```

```python
outlier_threshold(df, 'Age')
```
